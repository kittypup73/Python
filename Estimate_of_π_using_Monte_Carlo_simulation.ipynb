```python
#Determine the value of Pi using Monte Carlo simulations

#we import our libraries
import numpy as np
import matplotlib.pyplot as plt

#we define a variable as 10000, thus generating 10000 random points for the MC simuln
Nt = 10000

#we define two arrays of 10000 random  numbers between 0 and 1 (our range is [0,1))
x0 = np.random.random(Nt)
y0 = np.random.random(Nt)

#Goal: Simulate a unit circle inside a square. We want to randomly throw points into a square and see how many land inside the unit circle (radius 1).
#therefore, our points cannot be in a range of [0,1) since that would cover only one quadrant
#therefore, our range should be [-1,1), so all four quadrants can be covered, if x and y belong to them
#therefore we change the range
#we multiply x0 (or y0) by 2.0, therefore range becomes [0,2). After that we subtract 1.0 from the set. Therefore, range becomes [-1,1)
x = x0*2.0 - 1.0
y = y0*2.0 - 1.0

#we define r
r = np.sqrt(pow(x,2)+pow(y,2))

#we define a sum of all pts inside the circle; now we have two main variables for our ops ahead, Nt and Nb
inside_circle = r<= 1.0
Nb = np.sum(inside_circle)

#we give the formula, and the order to print the final value
pi_estimate = 4.0085*Nb/Nt
print("Estimated value of pi:", pi_estimate)

#we plot
plt.figure(figsize = (6,6))
plt.scatter(x[inside_circle], y[inside_circle], color = 'blue', label = 'Inside Circle', s=1)
plt.scatter(x[~inside_circle], y[~inside_circle], color = 'red', label = 'Outside Circle', s=1)
plt.title('Estimation of value of Pi using Monte Carlo simulation')
plt.legend()
plt.grid()
plt.show()
```
