```python
#PROJECTILE MOTION
#Your software company gave you a task to design a game in which you shoot monsters by controlling launch velocity v and launch angle theta of a projectile.
#Derive an equation for drawing the trajectory of the projectile as a function of v and theta (and initial coordinates)

import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0,2.04,100)
theta = np.radians(89)
u = 10
g = 9.8
x = (u*np.cos(theta)*t)
y = x * np.tan(theta) - (g * x**2) / (2 * u**2 * np.cos(theta)**2)

plt.subplot(311)
plt.plot(x,y, color = 'gold')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.title('The trajectory of a projectile')

plt.subplot(312)
plt.plot(t,x, color = 'green')
plt.xlabel('t')
plt.ylabel('x')
plt.grid()
plt.title('x position wrt time')

plt.subplot(313)
plt.plot(t,y, color = 'skyblue')
plt.xlabel('t')
plt.ylabel('y')
plt.grid()
plt.title('y position wrt time')

plt.tight_layout()
```
