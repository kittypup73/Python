```python
#Particle Decay

import numpy as np
import matplotlib.pyplot as plt

from google.colab import drive
drive.mount('/content/drive')
pB_path = '/content/drive/My Drive/P4P projects/pB_data.txt'
pB_data = np.loadtxt(pB_path, dtype = float)
print(pB_data[:10])

mA = 2.0
mB = 1.0
mC = np.sqrt(mA**2-2*mA*np.sqrt(mB**2+pB_data**2)+mB**2)

mC_mean = np.mean(mC)
mC_stdvn = np.std(mC, correction = 1)

plt.hist(mC, bins = 50, color = 'gold')
plt.title('Distribution of $mC$ values')
plt.xlabel('$mC$')
plt.ylabel('frequency')
plt.axvline(mC_mean, color = 'red', linestyle = 'dashed', linewidth = 1.0, label = f"Mean = {mC_mean:.3f}")
plt.axvline(mC_mean + mC_stdvn, color = 'coral', linestyle = 'dashed', linewidth = 1.0, label = f"Std Dvn = {mC_mean + mC_stdvn:.3f}")
plt.axvline(mC_mean - mC_stdvn, color = 'coral', linestyle = 'dashed', linewidth = 1.0, label = f"Std Dvn = {mC_mean - mC_stdvn:.3f}")
plt.legend()
plt.show()
```
